```python
# =======================
# TASK 1: Conceptual Understanding
# =======================
print("===== TASK 1: Conceptual Understanding =====\n")

print("Q1. Difference between 'Love' and 'love' in NLP:")
print("Answer: 'Love' with capital L is different from 'love' in NLP. Computers see them as different tokens. Lowercasing helps standardize words.\n")

print("Q2. What happens if stopwords are not removed?")
print("Answer: Stopwords like 'is', 'the', 'and' increase noise, inflate token count, and may reduce model performance.\n")

print("Q3. Two real-world scenarios where removing stopwords can be harmful:")
print("1. Sentiment analysis: Removing 'not' or 'no' can flip the meaning of the sentence.")
print("2. Search engines / question answering: Stopwords help preserve context in queries.\n")

print("Q4. Difference between stemming and lemmatization:")
print("Answer: Stemming crudely cuts words to roots (may produce non-words). Lemmatization converts words to their dictionary base form (more accurate).")
print("-"*50 + "\n")

# =======================
# TASK 2: Build Advanced Preprocessing Function
# =======================
print("===== TASK 2: Preprocessing Function =====\n")

import re
import string

def preprocess_text(text):
    if not text or text.isspace():
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Handle repeated characters (3 or more) → reduce to 2
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize
    tokens = text.split()

    # Remove very short tokens (length <= 2) except meaningful ones
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]

    # Reconstruct cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

print("Preprocessing function created successfully.\n")
print("-"*50 + "\n")

# =======================
# TASK 3: Stress Testing
# =======================
print("===== TASK 3: Stress Testing =====\n")

sample_texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in sample_texts:
    tokens, cleaned = preprocess_text(text)
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-"*50)

# =======================
# TASK 4: Token Analytics
# =======================
print("\n===== TASK 4: Token Analytics =====\n")

for text in sample_texts:
    tokens, cleaned = preprocess_text(text)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens)/total_tokens if total_tokens else 0
    print(f"Original: {text}")
    print(f"Total tokens: {total_tokens}, Unique tokens: {unique_tokens}, Average token length: {avg_len:.2f}")
    print("-"*50)

print("\nAnalysis Questions:")
print("Most noisy sentence likely: 'I absolutely looooved this product 😍😍' or 'Win $$$ now!!! Limited offer!!!'")
print("Sentence retaining most meaningful tokens: 'This is THE best course!!!' or 'I am not happy with this'\n")

# =======================
# TASK 5: Frequency Analysis
# =======================
print("===== TASK 5: Frequency Analysis =====\n")

from collections import Counter

all_tokens = []
for text in sample_texts:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)
print("Top 10 most frequent words:", counter.most_common(10))
print("Top 5 least frequent words:", counter.most_common()[-5:])
print("-"*50 + "\n")

# =======================
# TASK 6: Build Full Pipeline
# =======================
print("===== TASK 6: Full Pipeline =====\n")

def full_pipeline(text_list):
    tokens_list = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        tokens_list.append(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": tokens_list,
        "clean_sentences": clean_sentences
    }

pipeline_result = full_pipeline(sample_texts)
print("Pipeline Result:")
print(pipeline_result)
print("-"*50 + "\n")

# =======================
# TASK 7: Error Handling
# =======================
print("===== TASK 7: Error Handling =====\n")

test_cases = ["", "😂😂😂", "12345"]
for tc in test_cases:
    tokens, cleaned = preprocess_text(tc)
    print("Original:", tc, "=> Tokens:", tokens, "=> Cleaned:", cleaned)
```
