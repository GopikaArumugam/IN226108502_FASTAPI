```python
# ================================
# SENTIMENT ANALYSIS - FULL PIPELINE (ONE CELL)
# ================================

# 🔹 1. IMPORT LIBRARIES
import pandas as pd
import numpy as np
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Libraries imported successfully")


# 🔹 2. LOAD DATASET (ROBUST WAY)
df = pd.read_csv(
    "IMDB Dataset.csv",
    engine='python',
    on_bad_lines='skip',
    encoding='latin1',
    quoting=3
)

print("\nDataset loaded successfully")
print("Shape:", df.shape)
print(df.head())


# 🔹 3. DATA UNDERSTANDING
print("\nClass distribution BEFORE cleaning:")
print(df['sentiment'].value_counts())

print("\nMissing values:")
print(df.isnull().sum())


# 🔹 4. CLEAN SENTIMENT COLUMN
df['sentiment'] = df['sentiment'].astype(str).str.strip().str.lower()
df['sentiment'] = df['sentiment'].map({'positive':1, 'negative':0})
df = df.dropna(subset=['sentiment'])
df['sentiment'] = df['sentiment'].astype(int)

print("\nClass distribution AFTER cleaning:")
print(df['sentiment'].value_counts())


# 🔹 5. HANDLE MISSING TEXT
df['review'] = df['review'].fillna("")
print("\nMissing review values handled")


# 🔹 6. NLP PREPROCESSING SETUP
nltk.download('stopwords')

stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

print("\nNLP setup completed")


# 🔹 7. CLEANING FUNCTION
def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)

    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [stemmer.stem(word) for word in words]

    return " ".join(words)

print("Cleaning function ready")


# 🔹 8. APPLY PREPROCESSING
df['cleaned'] = df['review'].apply(clean_text)

print("\nSample cleaned data:")
print(df[['review', 'cleaned']].head())


# 🔹 9. FEATURE ENGINEERING (TF-IDF)
tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(df['cleaned']).toarray()
y = df['sentiment']

print("\nFeature engineering completed")
print("Feature matrix shape:", X.shape)


# 🔹 10. TRAIN-TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print("\nTrain-Test split done")
print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))


# 🔹 11. MODEL 1 - LOGISTIC REGRESSION
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("\nLogistic Regression trained successfully")


# 🔹 12. MODEL 2 - NAIVE BAYES
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

print("Naive Bayes trained successfully")


# 🔹 13. MODEL 3 - DECISION TREE
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

print("Decision Tree trained successfully")


# 🔹 14. EVALUATION FUNCTION
def evaluate(y_test, y_pred):
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))


# 🔹 15. EVALUATE MODELS
print("\n🔹 Logistic Regression Performance")
evaluate(y_test, y_pred_lr)

print("\n🔹 Naive Bayes Performance")
evaluate(y_test, y_pred_nb)

print("\n🔹 Decision Tree Performance")
evaluate(y_test, y_pred_dt)


# 🔹 16. COMPARISON TABLE
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Naive Bayes', 'Decision Tree'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_nb),
        accuracy_score(y_test, y_pred_dt)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_lr),
        f1_score(y_test, y_pred_nb),
        f1_score(y_test, y_pred_dt)
    ]
})

print("\nModel Comparison:")
print(results)


# 🔹 17. CONFUSION MATRIX
cm = confusion_matrix(y_test, y_pred_lr)
print("\nConfusion Matrix (Logistic Regression):")
print(cm)


# 🔹 18. FINAL SUMMARY
print("\nFINAL SUMMARY:")
print("""
1. Preprocessing (stopword removal + stemming) improved results.
2. TF-IDF performed better than basic text representation.
3. Logistic Regression gave best performance.
4. Naive Bayes was faster but slightly less accurate.
5. Decision Tree showed overfitting.

Conclusion:
TF-IDF + Logistic Regression is the best approach for sentiment analysis.
""")
```
